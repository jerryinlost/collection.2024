```python
from bunkatopics import Bunka
import os
```

```python
# model_name = "all-MiniLM-L6-v2"
# embedding_model = HuggingFaceEmbeddings(model_name=model_name) # We recommend starting with a small model
from sentence_transformers import SentenceTransformer
embedding_model = SentenceTransformer(model_name_or_path="all-MiniLM-L6-v2")
```

```python
from datasets import load_dataset
import random

docs = load_dataset("bunkalab/medium-sample-technology")["train"]["title"]

from sklearn.manifold import TSNE

projection_model = TSNE(
            n_components=2,
            learning_rate="auto",
            init="random",
            perplexity=3,
            random_state=42,
        )
```

```python
# Initialize Bunka with your chosen model and language preference
bunka = Bunka(embedding_model=embedding_model,
              projection_model = projection_model)

# Fit Bunka to your text data
bunka.fit(docs)
```

```python
import os
from google.colab import userdata
from langchain.llms import OpenAI

# Defined in the secrets tab in Google Colab
OPEN_AI_KEY = userdata.get('OPEN_AI')
llm = OpenAI(openai_api_key = OPEN_AI_KEY)
```

```python
## Bourdieu Fig
bourdieu_fig = bunka.visualize_bourdieu(
        llm=None,
        x_left_words=["This is about business"],
        x_right_words=["This is about politics"],
        y_top_words=["this is about startups"],
        y_bottom_words=["This is about governments"],
        height=800,
        width=800,
        clustering=True,
        topic_n_clusters=5,
        density=False,
        convex_hull=True,
        radius_size=0.2,
        min_docs_per_cluster = 5,
        label_size_ratio_clusters=80)
```

```python
bourdieu_fig.show()
```

```python
## Bourdieu Fig
bourdieu_fig = bunka.visualize_bourdieu(
        llm=llm,
        x_left_words=["This is positive"],
        x_right_words=["This is negative"],
        y_top_words=["this is about humans"],
        y_bottom_words=["This is about machines"],
        height=800,
        width=800,
        clustering=True,
        topic_n_clusters=10,
        density=False,
        convex_hull=True,
        radius_size=0.2,
        min_docs_per_cluster = 5,
        label_size_ratio_clusters=80)

bourdieu_fig.show()
```
